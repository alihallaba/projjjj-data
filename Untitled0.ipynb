```python
plt.figure(figsize=(8, 6))
sns.histplot(df['On average, how many hours per day do you spend using your smartphone for non-academic purposes (e.g., social media, entertainment, messaging)?'], bins=10, kde=True)
plt.title('Distribution of Smartphone Usage')
plt.xlabel('Hours per Day')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='On average, how many hours per day do you spend using your smartphone for non-academic purposes (e.g., social media, entertainment, messaging)?', data=df, palette='viridis')
plt.title('Frequency of Smartphone Notifications Interruptions')
plt.xlabel('Frequency')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Have you noticed a correlation between your smartphone usage patterns (e.g., excessive screen time, specific apps) and changes in your productivity levels?', data=df, palette='Set2')
plt.title('Noticed correlation between Smartphone Usage and Productivity')
plt.xlabel('Correlation')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Have you noticed a correlation between your smartphone usage patterns (e.g., excessive screen time, specific apps) and changes in your productivity levels?'], bins=len(df['Have you noticed a correlation between your smartphone usage patterns (e.g., excessive screen time, specific apps) and changes in your productivity levels?'].unique()), kde=False)
plt.title('Noticed correlation between Smartphone Usage and Productivity')
plt.xlabel('Correlation')
plt.ylabel('Frequency')
plt.show()
```
